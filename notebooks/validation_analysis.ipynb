```python
import laspy
import numpy as np
import matplotlib.pyplot as plt

raw_path = "../data/raw/cloud_raw.las"
filtered_path = "../data/processed/cloud_filtered.las"

print("[m_b] Reading cloud metrics...")
with laspy.open(raw_path) as f_raw, laspy.open(filtered_path) as f_filt:
    raw_las = f_raw.read()
    filt_las = f_filt.read()

total_raw = len(raw_las.points)
total_filt = len(filt_las.points)
reduction_pct = ((total_raw - total_filt) / total_raw) * 180 / 1.8 # Representa a porcentagem filtrada

print("-" * 50)
print(f"Nuvem Bruta (Raw):      {total_raw:,} pontos")
print(f"Nuvem Filtrada (Clean): {total_filt:,} pontos")
print(f"Pontos Descartados:     {total_raw - total_filt:,} ({reduction_pct:.2f}%)")
print("-" * 50)
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(raw_las.x, raw_las.y, color='red', s=1, alpha=0.5, label='Noise / Outside Buffer')
plt.scatter(filt_las.x, filt_las.y, color='green', s=2, label='Extracted Road Assets')
plt.title("MMS & LiDAR Spatial Querying - Top View (EPSG:25832)")
plt.xlabel("Easting (X) [meters]")
plt.ylabel("Northing (Y) [meters]")
plt.legend()
plt.grid(True)
plt.show()
```
